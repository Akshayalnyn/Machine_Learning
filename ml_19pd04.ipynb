```python
x = np.linspace(-1, 1, 10000)
y = np.exp(x) + np.abs(x)
plt.plot(x, y)
plt.show()

x = np.linspace(-1, 1, 10000)
y = -(np.exp(x) + np.abs(x))
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-1, 1, 10000)
y = np.minimum(np.abs(x), np.exp(x))
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('MINIMUM OF CONVEX FUNCTIONS (|x|, e^x)')
```

```python
x1 = np.linspace(-1, 1, 10000)
y1 = np.maximum(np.abs(x), np.exp(x))
plt.plot(x1, y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('MAXIMUM OF CONVEX FUNCTIONS (|x|, e^x)')
```

DOUBLE DIFFERENTIATING TO FIND IF IT IS CONCAVE OR CONVEX

```python
def isconvex(function):
  x = symbols('x')
  y = diff(function, x, 2)
  if(is_increasing(y, Interval(0, oo))):
    print("Convex");
  else:
    print("Concave");
```
